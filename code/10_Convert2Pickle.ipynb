```python
import pickle
import pandas as pd

data_dir = "./../data/TCGA-BRCA/"
mod = ["mRNA", "DNAm" , "RPPA"]
```

```python
for omic in mod : 
    expr = pd.read_csv(f"{data_dir}raw/{omic}_Expr.csv" , index_col = 0)
    meta = pd.read_csv(f"{data_dir}raw/{omic}_Meta.csv" , index_col = 0)
    

    with open(f"{data_dir}{omic}.pkl", "wb") as f:  # 'wb' = write binary
        pickle.dump({'expr' : expr , 'meta' : meta}, f, protocol=pickle.HIGHEST_PROTOCOL)
```
